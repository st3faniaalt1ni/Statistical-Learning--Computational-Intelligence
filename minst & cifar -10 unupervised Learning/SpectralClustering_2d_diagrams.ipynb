```
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import Isomap, LocallyLinearEmbedding, TSNE, SpectralEmbedding
from sklearn.cluster import SpectralClustering
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
```

```
from data import prepare_data
```

```
# DATASET = 'MNIST'
DATASET = 'CIFAR10'

SEED = 0
np.random.seed(SEED)  # to reproduce results (same randomness)

X_train, _, y_train, _ = prepare_data(DATASET, apply_pca=False, train_samples=17000, test_samples=1000)

colors = {
    0: 'red',
    1: 'blue',
    2: 'green',
    3: 'black',
    4: 'purple',
    5: 'orange',
    6: 'cyan',
    7: 'magenta',
    8: 'yellow',
    9: '#d182ad',

}
```

```
# ISOMAP
embedding_isomap = Isomap(n_components=2, n_jobs=2) #resulting data has 2 dimensions, 'components'
X_train_transformed_isomap = embedding_isomap.fit_transform(X_train) #fit model and transform
```

```
x_plot = X_train_transformed_isomap[0:17000, :][:, 0]
y_plot = X_train_transformed_isomap[0:17000, :][:, 1]

for index, (x, y) in enumerate(zip(x_plot, y_plot)):
    plt.scatter(x, y, c=colors[y_train[index]], s=3)
plt.show()
```

```
embedding_lle = LocallyLinearEmbedding(n_components=2, n_jobs=2) #resulting data has 2 dimensions, 'components'
X_train_transformed_lle = embedding_lle.fit_transform(X_train) #fit model and transform
```

```
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower

x_plot = X_train_transformed_lle[0:17000, :][:, 0]
y_plot = X_train_transformed_lle[0:17000, :][:, 1]

for index, (x, y) in enumerate(zip(x_plot, y_plot)):
    plt.scatter(x, y, c=colors[y_train[index]], s=3)
plt.show()
```

```
embedding_tsne = TSNE(n_components=2, n_jobs=2) #resulting data has 2 dimensions, 'components'
X_train_transformed_tsne = embedding_tsne.fit_transform(X_train) #fit model and transform
```

```
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower

x_plot = X_train_transformed_tsne[0:17000, :][:, 0]
y_plot = X_train_transformed_tsne[0:17000, :][:, 1]

for index, (x, y) in enumerate(zip(x_plot, y_plot)):
    plt.scatter(x, y, c=colors[y_train[index]], s=3)
plt.show()
```

```
embedding_se = SpectralEmbedding(n_components=2, n_jobs=2) #resulting data has 2 dimensions, 'components'
X_train_transformed_se = embedding_se.fit_transform(X_train) #fit model and transform
```

```
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower

x_plot = X_train_transformed_se[0:17000, :][:, 0]
y_plot = X_train_transformed_se[0:17000, :][:, 1]

for index, (x, y) in enumerate(zip(x_plot, y_plot)):
    plt.scatter(x, y, c=colors[y_train[index]], s=3)
plt.show()
```
